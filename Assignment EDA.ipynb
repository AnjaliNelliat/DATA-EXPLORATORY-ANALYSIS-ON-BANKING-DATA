```python
#### From the histogram plot, we could infer that middle age group are more in the dataset.
```

```python
def Age(x):
    'divide the age into four categories'
    if x < 18:
        return "Student"
    elif 18 <= x < 30:
        return "Adult"
    elif 30 <= x < 50:
        return "Middle-Aged"
    elif 50 <= x < 100:
        return "Senior"
    else:
        return "unknown"


df1_appdata['Age_Group'] = df1_appdata.DAYS_BIRTH.apply(lambda x: Age(x))
```
